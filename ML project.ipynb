```python
import gradio as gr
import joblib
import numpy as np

# Load your trained SVM model
model = joblib.load(r"C:\Users\hp\Downloads.pkl")

# Define the prediction function
def predict_class(id,news_url,title,tweet_ids):
    features = np.array([[id,news_url,title,tweet_ids]])
    prediction = model.predict(features)
    return f"Predicted Class: {prediction[0]}"

# Define Gradio interface
inputs = [
    gr.Number(label="id"),
    gr.Number(label="news_url"),
    gr.Number(label="title"),
    gr.Number(label="tweet_ids"),
   ]

output = gr.Textbox(label="Prediction")

gr.Interface(fn=predict_class, inputs=inputs, outputs=output, title="fake news detection").launch()
```

```python

from fastapi import FastAPI
import joblib
import numpy as np

# Load model
model = joblib.load(r"C:\Users\hp\Downloads.pkl")

app = FastAPI(title="Fake News Detection API")

@app.post("/predict")
def predict(id: float, news_url: float, title: float, tweet_ids: float):
    features = np.array([[id, news_url, title, tweet_ids]])
    prediction = model.predict(features)
    return {"Predicted Class": int(prediction[0])}
```

```python
from flask import Flask, render_template, request
import joblib
import numpy as np

app = Flask(__name__)

# Load trained model
model = joblib.load(r"C:\Users\hp\Downloads.pkl")

@app.route("/", methods=["GET", "POST"])
def predict():
    prediction = ""

    if request.method == "POST":
        id = float(request.form["id"])
        news_url = float(request.form["news_url"])
        title = float(request.form["title"])
        tweet_ids = float(request.form["tweet_ids"])

        features = np.array([[id, news_url, title, tweet_ids]])
        result = model.predict(features)

        prediction = f"Predicted Class: {result[0]}"

    return render_template("index.html", prediction=prediction)

if __name__ == "__main__":
    app.run(debug=True)
```
